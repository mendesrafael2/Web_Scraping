Atividade: 

A tarefa consiste em acessar o site mencionado e fazer uma raspagem dos links contidos nele. Salve os links raspados em uma lista de strings. Após ter uma lista transforme em um DataFrame de coluna única. Depois salve o DataFrame em um csv.

```python
# Abrir URLs
import requests as rq

# Navegar em tags HTML
from bs4 import BeautifulSoup 

import pandas as pd
```

## Acessar o site mencionado e fazer uma raspagem dos links contidos nele

```python
r = requests.get('http://200.152.38.155/CNPJ/')
soup = BeautifulSoup(r.text,'html.parser')

# soup.find_all('a') -> Encontra os dados que utilizam a tag <a>
links = soup.find_all('a')
```

## Salve os links raspados em uma lista de strings 
* Foram eleminados os 5 primeiros links

```python
# soup.find_all('a') -> Retorna o atributo "href" de cada uma das tags

lista_links = []

for link in links[5:]:
    lista_links.append(link.get('href'))
```

##  Transforme a lista em um DataFrame de coluna única

```python
df_links = pd.DataFrame(data=lista_links, columns=['Link'])
df_links
```

## Salve o DataFrame em um csv

```python
df_links.to_csv('links.csv',index=False)
```
